### **Consigna 10**

> * Considerando los vuelos correspondientes al tramo más frecuente y que no fueron cancelados, realizar un diagrama de cajas que permita ver la distribución de la duración de los vuelos.
>
> * Una vez hecho el diagrama, ¿a qué consideran que pueden deberse los valores extremos que se visualizan en él?

```python
# COMPLETAR.
diagrama_cajas_datos = vuelos.loc[
    (vuelos['Origin'] == 'SAN') & 
    (vuelos['Dest'] == 'LAX') & 
    (vuelos['ActualElapsedTime'] > 0) & 
    (vuelos['Cancelled'] == 0)
]

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.boxplot(diagrama_cajas_datos['ActualElapsedTime'])
plt.title('Distribución de la duración de vuelos SAN-LAX (no cancelados)')
plt.ylabel('Duración (minutos)')
plt.show()
```

### **Consigna 11**

> Hacer una figura que permita ver si la distribución de la duración de los vuelos considerados en la consigna anterior varía o no en función de si ocurre un día o de semana o de fin de semana y comentarla.

```python
# COMPLETAR.
plt.figure(figsize=(10, 6))
diagrama_cajas_datos['TipoDia'] = diagrama_cajas_datos['DayOfWeek'].apply(lambda x: 'Fin de semana' if x in [6, 7] else 'Semana')
diagrama_cajas_datos.boxplot(column='ActualElapsedTime', by='TipoDia')
plt.title('Duración de vuelos SAN-LAX según tipo de día')
plt.suptitle('')
plt.xlabel('Tipo de día')
plt.ylabel('Duración (minutos)')
plt.show()
```

### **Consigna 12**

> Hacer una figura que permita ver, por mes, cuántos vuelos tienen demora en la salida y cuántos, en el arribo y comentarla.

```python
# COMPLETAR.

demoras_por_mes = pd.DataFrame({
    'Demora en salida': vuelos[vuelos['DepDelay'] > 0].groupby('Month').size(),
    'Demora en arribo': vuelos[vuelos['ArrDelay'] > 0].groupby('Month').size()
})

demoras_por_mes.plot(kind='line', figsize=(10,6))
plt.title('Cantidad de vuelos con demora en salida y arribo por mes')
plt.xlabel('Mes')
plt.ylabel('Cantidad de vuelos')
plt.xticks(rotation=0)
plt.legend()
plt.show()
```
