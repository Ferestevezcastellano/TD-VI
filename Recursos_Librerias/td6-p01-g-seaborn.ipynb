```python
import seaborn as sns
```

Sigamos la misma lógica que en la *notebook* de Matplotlib.

```python
datos = sns.load_dataset(name = 'taxis')
```

Tal como hicimos con Matplotlib, primero, hagamos la versión por defecto y, luego, mejorémosla un poco.

```python
sns.scatterplot(data = datos, x = 'distance', y = 'fare')
```

```python
import matplotlib.pyplot as plt

sns.regplot(data = datos, x = 'distance', y = 'fare',
            scatter_kws = {'color': 'black', 'alpha': 0.2},
            line_kws = {'ls': '--', 'linewidth': 2, 'color': 'blue'},
            ci = None)
plt.title('¿A mayor distancia, mayor es la tarifa?')
plt.xlabel('Distancia (mi)')
plt.ylabel('Tarifa\n(USD)', rotation = 0, labelpad = 25)
plt.show()
```
