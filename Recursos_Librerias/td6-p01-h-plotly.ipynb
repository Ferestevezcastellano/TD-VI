# **Plotly**

Por convención, el alias elegido para Plotly (concretamente, para `plotly.express`) es `px`.

```python
import plotly.express as px
```

Sigamos la misma lógica que en las *notebooks* de Matplotlib y seaborn.

```python
import seaborn as sns
datos = sns.load_dataset(name = 'taxis')
```

Tal como hicimos con Matplotlib y seaborn, primero, hagamos la versión por defecto y, luego, mejorémosla un poco.

```python
px.scatter(data_frame = datos, x = 'distance', y = 'fare')
```

```python
px.scatter(data_frame = datos, x = 'distance', y = 'fare',
           color_discrete_sequence = ['black'], opacity = 0.2,
           trendline = 'ols', trendline_color_override = 'blue',
           title = '¿A mayor distancia, mayor es la tarifa?',
           labels = {'distance': 'Distancia (mi)', 'fare': 'Tarifa (USD)'},
           template = 'simple_white',
           width = 600) \
   .update_layout(title_x = 0.5)
```
